```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive

# Google Drive'ı /content/drive altına bağla
drive.mount('/content/drive')

# Görüntüyü yükle (renkli)
img_path = '/content/drive/MyDrive/GoruntuIsleme/images/cat.png'
img = cv2.imread(img_path)

# Görüntü başarıyla yüklenmediyse, uygun bir hata mesajı ver
if img is None:
    print("Görüntü yüklenemedi. Dosya adını ve yolunu kontrol edin.")
else:
    # Smoothing (Gaussian filtresi) işlemi
    smoothing_ratio = 10  # Smoothing oranı
    img_smoothed = cv2.GaussianBlur(img, (0, 0), smoothing_ratio)

    # Sharpening (Laplacian filtresi) işlemi
    sharpening_ratio = 1  # Sharpening oranı
    laplacian = cv2.Laplacian(img, cv2.CV_64F)
    img_sharpened = img - sharpening_ratio * laplacian

    # Görüntüleri göster
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Orijinal Görüntü')

    plt.subplot(1, 3, 2)
    plt.imshow(cv2.cvtColor(img_smoothed, cv2.COLOR_BGR2RGB))
    plt.title('Smoothing (Gaussian Filtre)')

    plt.subplot(1, 3, 3)
    plt.imshow(cv2.cvtColor(np.uint8(img_sharpened), cv2.COLOR_BGR2RGB))
    plt.title('Sharpening (Laplacian Filtre)')

    plt.show()
```
