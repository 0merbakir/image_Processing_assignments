```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive

# Google Drive'ı /content/drive altına bağla
drive.mount('/content/drive')

# Görüntüyü yükle (renkli)
img_path = '/content/drive/MyDrive/GoruntuIsleme/images/bird.png'
img = cv2.imread(img_path)

# Görüntü başarıyla yüklenmediyse, uygun bir hata mesajı ver
if img is None:
    print("Görüntü yüklenemedi. Dosya adını ve yolunu kontrol edin.")
else:
    # Sobel filtresi fonksiyonu (renkli görüntü için)
    def apply_sobel_filter_color(image):
        sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=1)
        sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=1)
        gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
        return gradient_magnitude

    # Görüntü boyutları
    height, width, channels = img.shape

    # Sobel filtresini her renk kanalına uygula
    img_sobel = apply_sobel_filter_color(img)

    # Görüntüleri göster
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(img)
    plt.title('Orijinal Görüntü')

    plt.subplot(1, 3, 2)
    plt.imshow(img_sobel)
    plt.title('Sobel Filtresi Sonucu')

    plt.show()
```
