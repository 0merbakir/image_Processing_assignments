```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive

# Google Drive'ı /content/drive altına bağla
drive.mount('/content/drive')

# Görüntüyü yükle (renkli)
img_path = '/content/drive/MyDrive/GoruntuIsleme/images/money.png'
img = cv2.imread(img_path)

# Görüntü başarıyla yüklenmediyse, uygun bir hata mesajı ver
if img is None:
    print("Görüntü yüklenemedi. Dosya adını ve yolunu kontrol edin.")
else:
    # Açma (opening) işlemi
    kernel_opening = np.ones((5, 5), np.uint8)
    img_opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel_opening)

    # Kapama (closing) işlemi
    kernel_closing = np.ones((5, 5), np.uint8)
    img_closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel_closing)

    # Görüntüleri göster
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Orijinal Görüntü')

    plt.subplot(1, 3, 2)
    plt.imshow(cv2.cvtColor(img_opening, cv2.COLOR_BGR2RGB))
    plt.title('Açma (Opening)')

    plt.subplot(1, 3, 3)
    plt.imshow(cv2.cvtColor(img_closing, cv2.COLOR_BGR2RGB))
    plt.title('Kapama (Closing)')

    plt.show()
```
