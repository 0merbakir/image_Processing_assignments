```python
import cv2
import numpy as np
from google.colab import drive
from matplotlib import pyplot as plt

# Google Drive'ı /content/drive altına bağla
drive.mount('/content/drive')

# Örnek bir renkli resim yükleyelim
image_path = '/content/drive/MyDrive/GoruntuIsleme/images/robert.jpg'
img = cv2.imread(image_path, cv2.IMREAD_COLOR)

# Görüntü özelliklerini yazdır
print("Görüntü Boyutu:", img.size)
print("Görüntü Veri Tipi:", img.dtype)
print("Görüntü Boyutları (Row, Column, Channels):", img.shape)

# Laplace filtresini uygula
laplacian_img = cv2.Laplacian(img, cv2.CV_64F)

# Orijinal ve Laplace filtresi uygulandıktan sonraki resimleri göster
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()

plt.imshow(laplacian_img)
plt.title('Laplace Sharpened Image')
plt.show()
```
