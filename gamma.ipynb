```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive

# Google Drive'ı /content/drive altına bağla
drive.mount('/content/drive')

# Görüntüyü yükle (renkli)
img_path = '/content/drive/MyDrive/GoruntuIsleme/images/haluk_bilginer.png'
img = cv2.imread(img_path)

# Görüntü başarıyla yüklenmediyse, uygun bir hata mesajı ver
if img is None:
    print("Görüntü yüklenemedi. Dosya adını ve yolunu kontrol edin.")
else:
    # Gamma dönüşümü fonksiyonu (renkli görüntü için)
    def gamma_transform_color(image, gamma):
        return np.clip((image / 255.0) ** gamma * 255, 0, 255).astype(np.uint8)

    # Görüntü boyutları
    height, width, channels = img.shape

    # Gamma dönüşümü uygula
    gamma_value = 1.5  # İstediğiniz gamma değerini seçebilirsiniz
    img_gamma_transformed = gamma_transform_color(img, gamma_value)

    # Görüntüleri göster
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Orijinal Görüntü')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(img_gamma_transformed, cv2.COLOR_BGR2RGB))
    plt.title(f'Gamma Dönüşümü (Gamma={gamma_value})')

    plt.show()
```
