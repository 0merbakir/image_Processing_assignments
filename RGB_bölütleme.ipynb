```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive

# Google Drive'ı /content/drive altına bağla
drive.mount('/content/drive')

# Görüntüyü yükle (renkli)
img_path = '/content/drive/MyDrive/GoruntuIsleme/images/fruits.png'
img = cv2.imread(img_path)

# Görüntü başarıyla yüklenmediyse, uygun bir hata mesajı ver
if img is None:
    print("Görüntü yüklenemedi. Dosya adını ve yolunu kontrol edin.")
else:
    # Renk kanallarına ayrılmış görüntü
    b, g, r = cv2.split(img)

    # Her bir renk kanalı için bölütlenme işlemi
    _, thresh_b = cv2.threshold(b, 128, 255, cv2.THRESH_BINARY)
    _, thresh_g = cv2.threshold(g, 128, 255, cv2.THRESH_BINARY)
    _, thresh_r = cv2.threshold(r, 128, 255, cv2.THRESH_BINARY)

    # Her bir renk kanalını birleştir
    img_thresholded = cv2.merge([thresh_b, thresh_g, thresh_r])

    # Görüntüleri göster
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title('Orijinal Görüntü')

    plt.subplot(1, 2, 2)
    plt.imshow(img_thresholded)
    plt.title('Bölütlenmiş Görüntü')

    plt.show()
```
